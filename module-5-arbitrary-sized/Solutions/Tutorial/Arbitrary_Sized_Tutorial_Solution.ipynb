```python
from cs103 import *
from typing import List

@typecheck
def any_positive(loi: List[int]) -> bool:
    """
    return True if any element in loi is positive
    """
    # return False #stub
    # template from List[int]
    
    res = False
    for i in loi:
        if i > 0:
            res = True
    return res
    
start_testing()

expect(any_positive([]), False)
expect(any_positive([1,2,3]), True)
expect(any_positive([1,-2,-7]), True)
expect(any_positive([-1,-92,10]), True)
expect(any_positive([-11,-9,-10]), False)

summary()
```

```python
from cs103 import *
from typing import List

@typecheck
def any_positive2(loi: List[int]) -> bool:
    """
    return True if any element in loi is positive
    """
    # return False #stub
    # template from List[int]
    
    for i in loi:
        if i > 0:
            return True
    return False
    
start_testing()

expect(any_positive2([]), False)
expect(any_positive2([1,2,3]), True)
expect(any_positive2([1,-2,-7]), True)
expect(any_positive2([-1,-92,10]), True)
expect(any_positive2([-11,-9,-10]), False)

summary()
```

## Problem 4

Design a function that takes a list of government grants and finds the name of the organization that received the smallest grant. If two or more are tied, return the first one.

Hints:

2. You may want to design your loop to find the *whole* grant with the smallest dollar value. Once you've done that, is there a way to get just the organization's name after the loop?

```python
@typecheck
def received_smallest_grant(logg: List[GovernmentGrant]) -> str:
    """
    return the name of the organization that received the smallest grant
    ASSUME: logg is not empty
    """
    #return "" #stub
    # template from List[GovernmentGrant]
    
    # acc contains the GovernmentGrant with the smallest grant seen so far
    acc = logg[0] # type: GovernmentGrant
    for gg in logg:
        if smaller_grant(gg, acc):
            acc = gg
    return acc.org

@typecheck
def smaller_grant(gg1: GovernmentGrant, gg2: GovernmentGrant) -> bool:
    """
    return True if gg1 has a smaller grant than gg2, False otherwise
    """
    #return False #stub
    # template from GovernmentGrant (all selectors from both inputs)
    return gg1.amt < gg2.amt


start_testing()

expect(received_smallest_grant([ENTROPY]), "Low Entropy Foundation")
expect(received_smallest_grant([FOLK_MUSIC, ENTROPY]), "Low Entropy Foundation")
expect(received_smallest_grant([FOLK_MUSIC, SPOLY, HERITAGE]), "Barriere and District Heritage Society")

summary()

start_testing()

expect(smaller_grant(SPOLY, HERITAGE), False)
expect(smaller_grant(HERITAGE, SPOLY), True)
expect(smaller_grant(ENTROPY, ENTROPY), False)

summary()
```
