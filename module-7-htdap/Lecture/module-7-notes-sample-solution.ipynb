```python
import csv

@typecheck
def read(filename: str) -> List[AirQuality]:
    """    
    reads information from the specified file and returns a list of air quality
    """
    # Template from HtDAP
    # loaq contains the result so far
    loaq = [] # type: List[AirQuality]

    with open(filename) as csvfile:
        
        reader = csv.reader(csvfile)
        next(reader) # skip header line

        for row in reader:
            # you may not need to store all the rows, and you may need
            # to convert some of the strings to other types
            aq = AirQuality(row[1], 
                            parse_province(row[4]),
                            parse_float(row[9]))
            loaq.append(aq)
    
    return loaq

@typecheck
def parse_province(p: str) -> Province:
    """
    return the Province represented by p
    """
    #return Province.BC
    # template inspired by Province
    if p == "British Columbia":
        return Province.BC
    elif p == "Alberta":
        return Province.AB   
    elif p == "Saskatchewan":
        return Province.SK  
    elif p == "Manitoba":
        return Province.MB 
    elif p == "Ontario":
        return Province.ON  
    elif p == "Quebec":
        return Province.QC  
    elif p == "Prince Edward Island":
        return Province.PE
    elif p == "Newfoundland and Labrador":
        return Province.NL 
    elif p == "Nova Scotia":
        return Province.NS  
    elif p == "New Brunswick":
        return Province.NB 
    elif p == "Yukon":
        return Province.YT  
    elif p == "Northwest Territories":
        return Province.NT 
    elif p == "Nunavut":
        return Province.NU   

start_testing()

# Examples and tests for read
expect(read('average_fine_particulate_matter-test1.csv'), [AQ1, AQ2])
expect(read('average_fine_particulate_matter-test2.csv'), 
       [AirQuality("NAPS HOUSE", Province.YT, None), 
        AirQuality("STEELE STREET", Province.YT, None)])

# Examples and tests for parse_province
expect(parse_province("British Columbia"), Province.BC)
expect(parse_province("Alberta"), Province.AB)
expect(parse_province("Saskatchewan"), Province.SK)
expect(parse_province("Manitoba"), Province.MB)
expect(parse_province("Ontario"), Province.ON)
expect(parse_province("Quebec"), Province.QC)
expect(parse_province("Prince Edward Island"), Province.PE)
expect(parse_province("Nova Scotia"), Province.NS)
expect(parse_province("New Brunswick"), Province.NB)
expect(parse_province("Newfoundland and Labrador"), Province.NL)
expect(parse_province("Yukon"), Province.YT)
expect(parse_province("Northwest Territories"), Province.NT)
expect(parse_province("Nunavut"), Province.NU)

summary()

```

```python
###########
# Functions

@typecheck
def main(filename: str, prov: str) -> Optional[float]:
    """
    Reads the file from given filename, then finds and returns the average 
    concentration of fine particulate matter for the given province
    
    returns None if there are no concentration readings for that province
    """
    # Template from HtDAP, based on function composition 
    return average_conc(read(filename), prov) 
    
@typecheck
def average_conc(loaq: List[AirQuality], prov: str) -> Optional[float]: 
    """ 
    finds and returns the average 
    concentration of fine particulate matter for the given province
    
    returns None if loaq doesn't have any readings for the given province
    """ 

    #return 0.0
    # template based on composition
    return average(filter_by_prov(loaq, prov))

@typecheck
def average(loaq: List[AirQuality]) -> Optional[float]:
    """
    returns the average concentration of fine particulate matter for air 
    quality readings in loaq. returns None if loaq is empty or if all
    air quality readings are missing
    """
    #return 0.0
    # template from List[AirQuality]

    # total contains the total sum of concentrations seen so far
    total = 0.0 # type: float
    # count contains the number of concentrations seen so far
    count = 0 # type: int
    
    
    for aq in loaq:
        if has_conc(aq):
            total = total + aq.conc
            count = count + 1
    if count == 0:
        return None
    
    return total/count
    

@typecheck
def filter_by_prov(loaq: List[AirQuality], prov: str) -> List[AirQuality]:
    """
    return a list of air quality readings from loaq that are in prov
    """
    #return []
    # template from List[AirQuality] with additional parameter
    # acc contains the result so far
    acc = [] # type: List[AirQuality]
    
    for aq in loaq:
        if same_prov(aq, parse_province(prov)):
            acc.append(aq)
    return acc

@typecheck
def same_prov(aq: AirQuality, p: Province) -> bool:
    """
    return True if aq is in province p, False otherwise
    """
    #return True
    # template from aq with additional parameter
    # DISCUSSION QUESTION: could/should we have created an additional 
    # helper function here?
    return aq.prov == p

@typecheck
def has_conc(aq: AirQuality) -> bool:
    """
    return True if aq has a concentration (ie, it is not missing)
    """
    #return True
    # template from AirQuality
    return not_missing(aq.conc)

@typecheck
def not_missing(c: Concentration) -> bool:
    """
    return False if c is None, True otherwise
    """
    #return False
    # template from Concentration
    if c == None:
        return False
    else:
        return True

start_testing()

# Examples and tests for main
expect(main("average_fine_particulate_matter-test1.csv", "British Columbia"), 5.4)
expect(main("average_fine_particulate_matter-test2.csv", "Yukon"), None)

# Examples and tests for average_conc 
expect(average_conc([], "British Columbia"), None) 
expect(average_conc([AQ1, AQ2], "Alberta"), None) 
expect(average_conc(L1, "British Columbia"), 5.4) 
expect(average_conc(L1, "Newfoundland and Labrador"), 4.7) 

# Examples and tests for average
expect(average([]), None) 
expect(average([AQ1, AQ2]), 5.4) 
expect(average([AQ3]), 4.7) 

# Examples and tests for filter_by_prov 
expect(filter_by_prov([], "British Columbia"), [])
expect(filter_by_prov(L1, "British Columbia"), [AQ1, AQ2])
expect(filter_by_prov(L1, "Newfoundland and Labrador"), [AQ3])

# Examples and tests for same_prov
expect(same_prov(AQ1, Province.BC), True)
expect(same_prov(AQ1, Province.AB), False)
expect(same_prov(AQ3, Province.BC), False)
expect(same_prov(AQ3, Province.NL), True)

# Examples and tests for has_conc
expect(has_conc(AQ1), True)
expect(has_conc(AQ4), False)

# Examples and tests for not_missing
expect(not_missing(1.2), True)
expect(not_missing(None), False)

summary()
```
